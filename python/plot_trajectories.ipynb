```python
from matplotlib import pyplot
%matplotlib inline
import numpy as np
from flowtracks.io import trajectories_ptvis
```

```python
inName = '../res/ptv_is.%d'
trajects = trajectories_ptvis(inName, traj_min_len=5, xuap=False)
```

```python
trajects
```

```python
pyplot.figure(figsize = (12, 10))

for traj in trajects:
	pyplot.plot(traj.pos()[:,0], traj.pos()[:,1], '.')

pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.show()
```

```python
pyplot.figure(figsize = (12, 10))

for traj in trajects:
	pyplot.plot(traj.pos()[:,0], traj.pos()[:,2], '.')

pyplot.xlabel('x')
pyplot.ylabel('z')
pyplot.show()
```

```python
pyplot.figure(figsize = (12, 10))

for traj in trajects:
	pyplot.plot(traj.pos()[:,1], traj.pos()[:,2], '.')

pyplot.xlabel('y (m)')
pyplot.ylabel('z (m)')
pyplot.show()
```
